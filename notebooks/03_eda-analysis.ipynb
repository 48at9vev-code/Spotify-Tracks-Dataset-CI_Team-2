## Spotify tracks dataset - ETL

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
```

```python
# Loading data from a CSV file
data = pd.read_csv('dataset.csv')
```

```python
# Checking the first few rows of the dataset
df = data
df
```

```python
# Droping 'Unnamed: 0' column
df = df.drop(columns=['Unnamed: 0'])
df
```

```python
df.info()
```

```python
# Checking the missing values in the dataset
df.isnull().sum()
```

```python
# investigate the missing values further
missing_values = df[df.isnull().any(axis=1)]
missing_values
```

```python
# The key data is missed in the row, so we will drop the row
df = df.dropna()
df.isnull().sum()
```

```python
# Investigating artists names for duplicates
df['artists'].value_counts()
```

```python
# Removing in the artists names' anomalies like extra spaces or ';' separators
df['artists'] = df['artists'].str.strip()
df['artists'] = df['artists'].str.replace(';', ', ')
df['artists'].value_counts()
```

```python
# Striping leading and trailing spaces from track_id, track_name, album_name, track_genre
df['track_id'] = df['track_id'].str.strip()
df['track_name'] = df['track_name'].str.strip()
df['album_name'] = df['album_name'].str.strip()
df['track_genre'] = df['track_genre'].str.strip()
```

```python
# Investigating track_id for duplicates
df['track_id'].value_counts()
```

```python
# Checking the rows with duplicate track_ids and sorting them by track_id
duplicate_track_ids = df[df.duplicated(subset=['track_id'], keep=False)]
duplicate_track_ids.sort_values(by='track_id')
```

```python
# Checking the rows with duplicate track_name
duplicate_track_names = df[df.duplicated(subset=['track_name'], keep=False)]
duplicate_track_names.value_counts()
```

```python
# Sorting duplicate by track_name for the further check
duplicate_track_ids.sort_values(by='track_name')
```

```python
# Deleting duplicate track_name rows with the same track_genre and track_id, keeping the first occurrence
df = df.drop_duplicates(subset=['track_name', 'track_genre', 'track_id'], keep='first')
```

```python
# Reseting df index
df = df.reset_index(drop=True)
```

```python
# For further investigation, saving the track_genre names to a separate CSV file
df_genre = df['track_genre'].unique()
df_genre_df = pd.DataFrame(df_genre, columns=['track_genre'])
df_genre_df.to_csv('track_genres.csv', index=False)
```

```python
# Adding a new column that combines similar genres into broader categories.
# Use a dictionary of genre strings to check the 'track_genre' column
common_strings = {
    'rock': 'rock',
    'grunge': 'rock',
    'alternative': 'rock',
    'pop': 'pop',
    'j-': 'jazz',
    'latin': 'latino',
    'latino': 'latino',
    'jazz': 'jazz',
    'hard': 'hard',
    'electro': 'electronic',
    'dance': 'dance',
    'metal': 'metal',
    'grindcore': 'metal',
    'songwriter': 'songwriter',
    'idm': 'dance',
    'house': 'house',
    'techno': 'techno',
    'punk': 'punk',
    'reggae': 'reggae',
    'blues': 'blues',
    'r-n-b': 'blues',
    'pagode': 'brazilian',
    'brazil': 'brazilian',
    'mpb': 'brazilian',
    'samba': 'brazilian',
    'forro': 'brazilian',
    'malay': 'malaysian',
    'kids': 'children',
    'children': 'children',
    'edm': 'electronic',
    'dubstep': 'electronic',
    'breakbeat': 'electronic',
    'honky-tonk': 'country'
}

df['track_genre_category'] = df['track_genre'].apply(lambda genre: 
                                                     next((value for key, value in common_strings.items() if key in genre.lower()), genre))
df['track_genre_category'].value_counts()
```

## Insights and Hypotheses

### Exploratory Insights

- The distribution of **track popularity** is **imbalanced**, **highly multimodal**, and **right-skewed**. So Spearman correlation  is more useful for this dataset than Pearson.
- The largest number of tracks in the dataset belongs to the following genre categories: **rock, pop, metal, Brazilian, and electronic**.
- The majority of tracks have a **danceability around 0.6**, show a **trend toward high energy**, and exhibit **high loudness (~0.8)**, likely reflecting **heavy dynamic compression during mastering**. Most tracks are **music rather than speech**, are **non-acoustic (synth-based)**, **contain vocals**, and are **studio recordings rather than live performances**. The typical tempo ranges between **120–130 BPM**.
- The **highest median popularity** is observed among tracks in the following genres: **hip-hop, chill, sad, British, emo, ambient, anime, and piano**.
- The genres with the **highest hit rate** (more than **10% of tracks reaching popularity > 70**) are **indie, reggae, hip-hop, Latino, and soul**. An additional **8–10% of tracks** in **rock and pop, dance, electronic, and soul** reach popularity above 70.
- Tracks with the **highest average popularity** are **neither highly danceable nor loud**, but are more likely to be **live recordings**.
- Most songs are in **major mode**, but **major and minor modes show similar popularity levels**.
- Tracks with tempos between **60–200 BPM** show **approximately the same popularity**, with a slight drop **below 60 BPM and above 200 BPM**; however, such tempos are **rare in music**.

## Popularity Drivers (Core Hypotheses)

### H1: More energetic tracks are more popular
- **Test:** Spearman correlation  
- **Result:** **Rejected**  
- Popularity varies by less than **~5 points** across the full energy range, with the highest popularity observed around **energy ≈ 0.6**.

### H2: Louder tracks have higher popularity
- **Test:** Scatter plot + LOWESS (loudness vs. popularity)  
- **Result:** **Rejected**

### H3: Instrumental tracks are less popular
- **Test:** Scatter plot + LOWESS (instrumentalness vs. popularity)  
- **Result:** **Rejected**

---

## Musical Structure Hypotheses

### H4: Major-key songs are more popular than minor-key songs
- **Test:** Bar chart of mean popularity  
- **Result:** **Rejected** — both modes show similar popularity.

### H5: Extremely fast or slow tempos reduce popularity
- **Test:** Scatter plot + LOWESS (tempo vs. popularity); popularity by tempo bins  
- **Result:** **Partially supported** — popularity drops slightly (**1–3 points**) at extreme tempos.

### H7: High-valence (happier) tracks are more popular
- **Test:** Spearman correlation  
- **Result:** **Rejected** — highest popularity occurs around **valence ≈ 0.3**.

---

## Genre Effects

### H8: Popularity drivers differ by genre category
- **Test:** Spearman correlations per genre  
- **Result:** **Supported**  
- Feature–popularity relationships vary substantially across genres, indicating that feature-level analysis without genre grouping is misleading.

## Acoustic Properties

### H9: Acoustic tracks are less popular on average than non-acoustic tracks
- **Test:** Boxplot of popularity by acousticness bins  
- **Result:** **Supported**, though the effect size is small. The most popular tracks tend to be **hybrids**, combining acoustic and synthetic elements.
